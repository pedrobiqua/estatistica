# Visualização de Dados

```python
# Usando visualizações direto de dataframe do Pandas, usando numpy para gerar dados aleatórios

import numpy as np
import pandas as pd
```

```python
# Criando um dataframe vazio

df = pd.DataFrame()
```

```python
# Gráfico em linha

df = pd.DataFrame(np.random.randn(500), columns=["B"]).cumsum()
df["A"] = pd.Series(list(range(len(df))))
df.plot(x="A", y="B"); # df.plot.line(x="A", y="B");
```

```python
# Gráfico de área

df = pd.DataFrame(np.random.rand(20, 4),
       columns =['A', 'B', 'C', 'D'])
df.plot.area();
```

```python
# Gráfico de barras

df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
df.sum().plot.bar();
```

```python
# Gráfico de barras horizontais

df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
df.sum().plot.barh();
```

```python
# Histograma

df = pd.DataFrame(
        {
            "a": np.sqrt(np.random.randn(1000) + 1),
            "b": np.random.randn(1000),
        },
        columns=["a", "b", "c"],
    )
df.plot.hist(alpha=0.5);
```

```python
# Gráfico de dispersão

df = pd.DataFrame(np.random.rand(100, 2),
           columns =['a', 'b'])
df.plot.scatter(x ='a', y ='b');
```

```python
# Gráfico de caixa

df = pd.DataFrame(np.random.rand(50, 3), columns=["A", "B", "C"])
df.plot.box();
```

```python
# Gráfico de pizza

series = pd.Series(3 * np.random.rand(4), index=["a", "b", "c", "d"], name="Pie plot")
series.plot.pie(figsize=(6, 6));
```
